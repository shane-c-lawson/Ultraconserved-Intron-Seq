```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import mannwhitneyu
import statannot
from statannotations.Annotator import Annotator
from matplotlib.patches import PathPatch

table = pd.read_csv("psi_box_full.csv")
fig, ax = plt.subplots()
sns.set_style(rc= {'xtick.color': 'white'})#, 'figure.facecolor':'white'})
flierprops = dict(marker='o', markerfacecolor='black', markersize=1,  markeredgecolor='black')
palette = {'Exclusion':"seagreen",
			'Inclusion non-coding':"indianred",
			'Inclusion Coding':"lightgoldenrodyellow"}
b = sns.boxplot(data=table[["Inclusion non-coding", "Inclusion Coding", "Exclusion"]], ax=ax, palette = palette, showfliers = True, flierprops=flierprops, width=0.25)#, scale='width', zorder=3)

sns.despine()
sns.set(font="Arial")
ax.set_axisbelow(True)
b.set_ylabel("average PSI", fontsize=15)#, fontname="Times New Roman")
b.tick_params(labelsize=10)
newnames = ["Cons_NMD_In (Non-coding)", "Cons_NMD_In (Coding)", "Cons_NMD_Ex"]
b.set_xticklabels(newnames, size = 7.5)#, fontname="Times New Roman")#b.get_xticklabels(), size=5)#, rotation=45)#, size=10)

annotator = Annotator(ax, pairs=[("Inclusion Coding", "Inclusion non-coding"), ("Inclusion Coding", "Exclusion"), ("Inclusion non-coding", "Exclusion")], data=table[["Inclusion non-coding", "Inclusion Coding", "Exclusion"]])
annotator.configure(test='t-test_ind', text_format='star', loc='outside', line_width = 0.5, line_offset = 0.5, alpha = 0.5)
annotator.apply_and_annotate()


def adjust_box_widths(g, fac):
    """
    Adjust the widths of a seaborn-generated boxplot.
    """
    # iterating through Axes instances
    for ax in g.axes:
        # iterating through axes artists:
        for c in ax.get_children():

            # searching for PathPatches
            if isinstance(c, PathPatch):
                # getting current width of box:
                p = c.get_path()
                verts = p.vertices
                verts_sub = verts[:-1]
                xmin = np.min(verts_sub[:, 0])
                xmax = np.max(verts_sub[:, 0])
                xmid = 0.5*(xmin+xmax)
                xhalf = 0.5*(xmax - xmin)

                # setting new width of box
                xmin_new = xmid-fac*xhalf
                xmax_new = xmid+fac*xhalf
                verts_sub[verts_sub[:, 0] == xmin, 0] = xmin_new
                verts_sub[verts_sub[:, 0] == xmax, 0] = xmax_new

                # setting new width of median line
                for l in ax.lines:
                    if np.all(l.get_xdata() == [xmin, xmax]):
                        l.set_xdata([xmin_new, xmax_new])


adjust_box_widths(fig, 1.75)

inc = table.loc[:,'Inclusion non-coding']
exc = table.loc[:,'Exclusion']
cod = table.loc[:,'Inclusion Coding']

test = (sum(cod) / len(cod))
test2 = (sum(inc) / len(inc))

t3, p3 = stats.ttest_ind(cod.dropna(), inc.dropna())# method = 'exact')
if (sum(inc.dropna()) / len(inc.dropna())) > test :
	sign = '+'
else :
	sign = '-'
print('Non vs. Cod: ' + sign + str(p3/2))


t4, p4 = stats.ttest_ind(cod.dropna(), exc.dropna())# method = 'exact')
if (sum(exc.dropna()) / len(exc.dropna())) > test :
	sign = '+'
else :
	sign = '-'
print('Cod vs. Exclusion: ' + sign + str(p4/2))


t5, p5 = stats.ttest_ind(inc.dropna(), exc.dropna())# method = 'exact')
if (sum(exc.dropna()) / len(exc.dropna())) > test2 :
	sign = '+'
else :
	sign = '-'
print('Non vs. Exclusion: ' + sign + str(p5/2))
```
